### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "resources/city_data.csv"
ride_data_to_load = "resources/ride_data.csv"

# Read the City 
city_data_df = pd.read_csv(city_data_to_load)

# and Ride Data
ride_data_df = pd.read_csv(ride_data_to_load)

city_data_df.head()
```

```python
city_data_df.head()
```

```python
ride_data_df.head()
```

### Merge the DataFrames

```python
# Combine the data into a single dataset - pyber_data_df
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type - rides_by_city_type
rides_by_city_type = pyber_data_df.groupby("type").count().ride_id

rides_by_city_type
```

```python
# 2. Get the total drivers for each city type - drivers_by_city_type
drivers_by_city_type = city_data_df.groupby("type").sum().driver_count

drivers_by_city_type
```

```python
#  3. Get the total amount of fares for each city type - fares_by_city_type
fares_by_city_type = pyber_data_df.groupby("type").sum().fare

fares_by_city_type
```

```python
#  4. Get the average (mean) fare per ride for each city type - mean_fare_by_city_type
mean_fare_by_city_type = pyber_data_df.groupby("type").mean().fare

mean_fare_by_city_type
```

```python
# 5. Get the average (mean) fare per driver for each city type. - mean_driver_fare_by_city_type
mean_driver_fare_by_city_type = pyber_data_df.groupby("type").sum().fare / drivers_by_city_type

mean_driver_fare_by_city_type
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.DataFrame({   "Total Rides" : rides_by_city_type, 
                                    "Total Drivers" : drivers_by_city_type, 
                                    "Total Fares" : fares_by_city_type, 
                                    "Average Fare per Ride" : mean_fare_by_city_type, 
                                    "Average Fare per Driver" : mean_driver_fare_by_city_type})
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df
```

```python
#  8. Format the columns.
pyber_summary_df["Total Fares"] = pyber_summary_df["Total Fares"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Ride"] = pyber_summary_df["Average Fare per Ride"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Driver"] = pyber_summary_df["Average Fare per Driver"].map("${:,.2f}".format)

pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df.head()
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date - fare_by_city_date_df

fare_by_city_date_df = pyber_data_df.groupby(["type","date"]).sum().fare.to_frame()

fare_by_city_date_df.head()
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the "pivot()" function.
# df = df.reset_index()

fare_by_city_date_df = fare_by_city_date_df.reset_index()
 
fare_by_city_date_df.head()
```

```python
# 4. Create a pivot table with the "date" as the index, the columns ="type", and values="fare" 
# to get the total fares for each type of city by the date. 

fare_by_city_date_df = fare_by_city_date_df.pivot(index="date",columns="type", values="fare")

fare_by_city_date_df.head()
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, "2019-01-01":"2019-04-29" - fare_dates_df

fare_dates_df = fare_by_city_date_df.loc["2019-01-01":"2019-04-29"]

fare_dates_df
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)

type(fare_dates_df.index)

fare_dates_df.index = pd.to_datetime(fare_dates_df.index)
fare_dates_df.head()
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
fare_dates_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week "W" and get the sum of the fares for each week - fare_dates_df

fare_dates_df = fare_dates_df.resample("W").sum()
fare_dates_df
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use("fivethirtyeight")

fare_dates_df.plot(figsize=(12,6),)
plt.xlabel("")
plt.ylabel("Fare ($USD)")
plt.title("Total Fare by City Type")
plt.savefig("analysis/Pyber_fare_summary.png")
```
